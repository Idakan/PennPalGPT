```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
```

```python
df = pd.read_csv('data_professors.csv')
```

```python
df = df.fillna(0)
```

```python
df.head(2)
```

```python
x = df.drop(columns=['professor'])
```

```python
model = KMeans(n_clusters=6)
model.fit(x)
```

```python
labels = model.predict(x)
labels
```

```python
df['label'] = labels
```
